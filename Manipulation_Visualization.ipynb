```python
df = pd.read_csv('Budget_of_Department_of_Space-cleaned')
```

```python
df_GDPgrowth = df.iloc[:, :2]
```

```python
df_GDPgrowth
```

```python
import matplotlib.pyplot as plt

# Assuming df_first_two_columns is your DataFrame with two columns

# Extract x and y data from the DataFrame
x = df_GDPgrowth['Calendar Year']
y = df_GDPgrowth['GDP (2011-12 base year) in crores(₹)']

# Create a line graph with customizations
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(x, y, marker='o', linestyle='-', color='b', label='GDP')
plt.title('GDP Over Time')
plt.xlabel('Year')
plt.ylabel('GDP (in crores ₹)')
plt.legend()
plt.grid(True)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Mark the sudden dip at 2019-20 as "COVID" (if it exists)
covid_year = '2019–20'
covid_data = df_GDPgrowth[df_GDPgrowth['Calendar Year'] == covid_year]

if not covid_data.empty:
    covid_gdp = covid_data['GDP (2011-12 base year) in crores(₹)'].values[0]
    plt.annotate('COVID', xy=(covid_year, covid_gdp), xytext=(covid_year, covid_gdp - 5000),
                 arrowprops=dict(arrowstyle='->'))

# Display the graph
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df_cleaned is your original DataFrame

# Extract the data for the two line graphs
x = df['Calendar Year']
y1 = df['Total Expenditure in crores (₹)']
y2 = df['Nominal INR (crore)-DOP']

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the Total Expenditure line graph
axes[0].plot(x, y1, marker='o', linestyle='-', color='b', label='Total Expenditure')
axes[0].set_title('Total Expenditure')
axes[0].set_xlabel('Calendar Year')
axes[0].set_ylabel('Total Expenditure (in crores ₹)')
axes[0].grid(True)
axes[0].tick_params(axis='x', rotation=45)
axes[0].xaxis.set_major_locator(plt.MaxNLocator(6))  # Display a limited number of x-axis labels
axes[0].legend()

# Plot the Nominal INR line graph
axes[1].plot(x, y2, marker='o', linestyle='-', color='g', label='Nominal INR (crore)-DOP')
axes[1].set_title('Nominal INR (crore)-DOP')
axes[1].set_xlabel('Calendar Year')
axes[1].set_ylabel('Nominal INR (in crores ₹)')
axes[1].grid(True)
axes[1].tick_params(axis='x', rotation=45)
axes[1].xaxis.set_major_locator(plt.MaxNLocator(6))  # Display a limited number of x-axis labels
axes[1].legend()

# Set a common title for both subplots
plt.suptitle('Two Line Graphs')

# Adjust spacing between subplots
plt.tight_layout(w_pad=4)

# Display the subplots
plt.show()
```
