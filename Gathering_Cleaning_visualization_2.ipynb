```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert numerical columns to appropriate data types
df["marketcap"] = df["marketcap"].astype(int)
df["price (USD)"] = df["price (USD)"].astype(float)
```

```python
# Plot 1: Bar Plot of the Top 10 Companies by Market Cap
top_10_marketcap = df.nlargest(10, "marketcap")
plt.figure(figsize=(12, 6))
sns.barplot(x="marketcap", y="Name", data=top_10_marketcap)
plt.xlabel("Market Cap (USD)")
plt.title("Top 10 Companies by Market Cap")
plt.show()
```

```python
df["marketcap"] = df["marketcap"].astype(int)
df["price (USD)"] = df["price (USD)"].astype(float)

# Group data by country and calculate the total market cap
country_marketcap = df.groupby("country")["marketcap"].sum()

# Sort countries by total market cap in descending order
country_marketcap = country_marketcap.sort_values(ascending=False)

# Plot 2: Improved Pie Chart of Market Cap Distribution by Country
plt.figure(figsize=(8, 8))

# Explode the largest slice
explode = (0.1, 0, 0, 0, 0)

# Define colors for the pie slices
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

plt.pie(
    country_marketcap,
    labels=country_marketcap.index,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.85,
    counterclock=False,
    colors=colors,
    explode=explode,
)

# Draw a circle at the center to make it a donut chart
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.title("Market Cap Distribution by Country", fontsize=16)

plt.show()
```

```python
# Plot 3: Scatter Plot of Market Cap vs. Price
plt.figure(figsize=(8, 6))
sns.scatterplot(x="marketcap", y="price (USD)", data=df)
plt.xlabel("Market Cap (USD)")
plt.ylabel("Price (USD)")
plt.title("Market Cap vs. Price")
plt.show()
```

```python
# Plot 4: Histogram of Price Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["price (USD)"], bins=20, kde=True)
plt.xlabel("Price (USD)")
plt.title("Price Distribution")
plt.show()
```

```python
# Plot 5: Box Plot of Price Distribution by Country
plt.figure(figsize=(10, 6))
sns.boxplot(x="country", y="price (USD)", data=df)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Price (USD)")
plt.title("Price Distribution by Country")
plt.show()
```

```python
# Plot 6: Bar Plot of the Top 10 Companies by Price
top_10_price = df.nlargest(10, "price (USD)")
plt.figure(figsize=(12, 6))
sns.barplot(x="price (USD)", y="Name", data=top_10_price)
plt.xlabel("Price (USD)")
plt.title("Top 10 Companies by Price")
plt.show()
```

```python
# Plot 7: Pie Chart of Company Count by Country
company_count_by_country = df["country"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(company_count_by_country, labels=company_count_by_country.index, autopct='%1.1f%%')
plt.title("Company Count by Country")
plt.show()
```

```python
# Plot 8: Bar Plot of Top 10 Companies by Rank
top_10_rank = df.nsmallest(10, "Rank")
plt.figure(figsize=(12, 6))
sns.barplot(x="Rank", y="Name", data=top_10_rank)
plt.xlabel("Rank")
plt.title("Top 10 Companies by Rank")
plt.show()
```
