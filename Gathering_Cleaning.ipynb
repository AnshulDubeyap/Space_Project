```python
##Webscraping Data Wikipidia
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Space_industry_of_India'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
data_table = soup.find_all('table')[0] 
columns = []
header_row = data_table.find('tr')
for header_cell in header_row.find_all('th'):
    column_name = header_cell.text.strip()
    columns.append(column_name)
data_dict = {column: [] for column in columns}    
for row in data_table.find_all('tr')[1:]:  # Skip the header row
    cells = row.find_all('td')
    if len(cells) == len(columns):
        for i, cell in enumerate(cells):
            data_dict[columns[i]].append(cell.text.strip())
for column in columns:
    print(column, data_dict[column])
else:
    print("Failed to retrieve the Wikipedia page.")
```

```python
df1 = pd.DataFrame(data_dict)
df1.head(30)
```

```python
df1.to_csv('List_of_notable_companies', index=False)
```

```python
url ='https://en.wikipedia.org/wiki/ISRO'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find all tables with the class "wikitable"
tables = soup.find_all("table", {"class": "wikitable"})

# Check if there are at least 12 tables
if len(tables) >= 12:
    # Get the 12th table
    table = tables[11]  # Note that Python uses 0-based indexing

    # Extract data from the table
    table_data = []
    for row in table.find_all("tr"):
        columns = row.find_all(["th", "td"])
        row_data = [column.get_text(strip=True) for column in columns]
        table_data.append(row_data)

    # Print the table data
    for row in table_data:
        print("\t".join(row))
else:
    print("Not enough tables found on the page.")
```

```python
df3 = pd.DataFrame(table_data)
df3.head(30)
```

```python
df3.to_csv('Budget_of_Department_of_Space', index=False)
```
